```python
data1=data[['heart_disease_present',
       'slope_of_peak_exercise_st_segment', 'resting_blood_pressure',
       'chest_pain_type', 'num_major_vessels',
       'fasting_blood_sugar_gt_120_mg_per_dl', 'resting_ekg_results',
       'serum_cholesterol_mg_per_dl', 'oldpeak_eq_st_depression', 'sex', 'age',
       'max_heart_rate_achieved', 'exercise_induced_angina']]
data1
```

```python
data1.info()
```

```python
plt.figure(figsize=(20,25), facecolor='white')#defining  canvas size
plotnumber = 1 #maintian count for graph

for column in data1:#accessing the columns
    if plotnumber<=13 :# as there are 15 columns in the data
        ax = plt.subplot(6,7,plotnumber)# plotting 13 graphs (6-rows,7-columns) ,plotnumber is for count 
        sns.distplot(data1[column])#plotting dist plot to know distribution
        plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
## Note:on distribution curve
#     Normal distribution: heart disease present, sexercise st segment,BP,EKG results,Cholestrol,sex,age,heart rate,exercise induced - Mean
#     
#     Not-normal distibution:chest pain,major vessels,sugar,depression - Median    
```

```python
# # Checking outliers

# In[29]:


plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1
for column in data1:
    if plotnumber<=13:
        ax=plt.subplot(6,7,plotnumber)
        sns.boxplot(x=data1[column],hue=data1.heart_disease_present)
        plt.xlabel(column,fontsize=10)
        plt.ylabel('heart_disease_present',fontsize=20)
    plotnumber+=1
plt.show()
plt.tight_layout()
```

```python
data1.columns


# In[32]:


data1['resting_blood_pressure']=data1['resting_blood_pressure'].replace(0,data1['resting_blood_pressure'].mean())
data1['chest_pain_type']=data1['chest_pain_type'].replace(0,data1['resting_blood_pressure'].median())
data1['num_major_vessels']=data1['num_major_vessels'].replace(0,data1['num_major_vessels'].median())
data1['fasting_blood_sugar_gt_120_mg_per_dl']=data1['fasting_blood_sugar_gt_120_mg_per_dl'].replace(0,data1['fasting_blood_sugar_gt_120_mg_per_dl'].median())
data1['serum_cholesterol_mg_per_dl']=data1['serum_cholesterol_mg_per_dl'].replace(0,data1['serum_cholesterol_mg_per_dl'].mean())
data1['oldpeak_eq_st_depression']=data1['oldpeak_eq_st_depression'].replace(0,data1['oldpeak_eq_st_depression'].median())
```
